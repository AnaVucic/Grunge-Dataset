```python
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv('nirvana_v2.csv')
wiki = pd.read_csv('nirvana_wiki.csv', dtype={'release_year': 'int32'})
```

```python
base.shape
```

```python
wiki.shape
```

```python
nirvana = base.merge(wiki, how='left', left_on=['track_name'], right_on=['songs'])
nirvana.drop('songs', axis=1,inplace=True)
```

```python
nirvana.shape
```

```python
nirvana[['track_name', 'album_name', 'original_release', 'writers', 'producers', 'release_year_y', 'release_year_x']].head(10)
```

```python
nirvana.to_csv('nirvana_v3.csv', index = False)
```

```python
list = []
for row in range(nirvana.shape[0]):
    if nirvana.release_year_x[row] != nirvana.release_year_y[row]:
        print(nirvana.release_year_x[row])
        print(nirvana.release_year_y[row])
        print()
print(sum(list))
```
